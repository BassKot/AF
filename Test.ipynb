```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('task2_1_',sep='\t',header=None)
```

Посмотрим на данные

```python
data.shape
```

Увидел, что много '-', значит надо заменить это

```python
data.replace('-',0, inplace=True)
data.replace('0',0,inplace=True)
data.head(10)
```

```python
data.describe()
```

Посчитаем количество уникальных значений в каждом столбце

```python
for col in data.columns:
    print(col,data[col].nunique())
```

Итак, у нас колонки с одним значением: 2,10,9,21,20,24,23,31,78-80,100-102,139,159,170,174
Которое очевидно не несет смысла, поэтому можно удалить стобцы

```python
data1 = data.loc[:,(data != 0).any(axis=0)]
data1.head()
```

```python
data1.shape
```

Немного сократили этим размерность признаков.
Далее: так как первая колонка похожа на хэш от пользователя, мы ее пока трогать не будем и все дальнейшие операции проведем над числовыми признаками

```python
from sklearn.cluster import KMeans
X = data1.drop([0],axis=1)
X.head()
```

Так как масштаб чисел в колонках разный, можно все отшкалировать. Плюс методом ближайших соседей попробуем сразу разбить на 2 класса

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=2,random_state=42)
y = km.fit_predict(X_scaled)

plt.scatter(X_scaled[:,0],X_scaled[:,1],c=y)
plt.show()
```

```python
plt.scatter(X_scaled[:,0],X_scaled[:,1],c=y)
plt.show()
```

Размерность признакового пространства = 170, в голове представить сложно. Значить, чтобы визуализировать все это добро, применим TSNE

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, init='random', random_state=0)
X_t = tsne.fit_transform(X_scaled)
print("Количество компонент: 2")
plt.scatter(X_t[:,0],X_t[:,1])
plt.show()
    
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 2, init='k-means++', n_jobs=-1)
data1['Possible labels'] = km.fit_predict(X)
plt.figure(figsize=(12,10))
plt.scatter(X_t[:, 0], X_t[:, 1], c=data1['Possible labels'], 
            edgecolor='none', alpha=0.7, s=40,
            cmap=plt.cm.get_cmap('nipy_spectral', 13))
plt.colorbar()
plt.title('T-SNE projection_2_clusters by KMeans')
```

Теперь Попробуем подстроить модель леса под значения, предсказанные KMeans, чтобы вытащить признаки, которые больше всего влияют на класс пользователя

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
y = data1['Possible labels']
feats = data1.drop(['Possible labels',0],axis=1)
rf = RandomForestClassifier()
rf.fit(feats,y)
```

Получим признаки с наибольшей значимостью

```python
print(sorted(rf.feature_importances_)[160:])
```

А теперь попробуем методом PCA сократить размерность признаков.

```python
from sklearn.decomposition import PCA

for i in range(2,15):
    pca = PCA(n_components=i)
    pca.fit(feats)
    print("-------------Количество компонент:",i)
    print("Explaned variance ratio:",pca.explained_variance_ratio_)
    print("CUMSUM: ",pca.explained_variance_ratio_.cumsum())
```

Как можно заметить, методом PCA достаточно будет перейти из 170 признаков в 4

```python
a = []                                         ### Записываем номера признаков, 
for i in range(len(rf.feature_importances_)):   # дающих наибольший вклад
    if rf.feature_importances_[i] > 0.02:
        a.append(i)
        
print(a)
```

Таким образом получили номера признаков, дающих наибольший вклад в определение класса "по версии" RFC
