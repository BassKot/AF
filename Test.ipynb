```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('task2_1_',sep='\t',header=None)
```

```python
data.replace('-',0, inplace=True)
data.head(10)
```

```python
data.describe()
```

Итак, у нас колонки с одним значением: 2,10,9,21,20,24,23,31,78-80,100-102,139,159,170,174
Которое очевидно не несет смысла, поэтому можно удалить стобцы

```python
data1 = data.loc[:,(data != 0).any(axis=0)]
data1.head()
```

```python
data1.shape
```

```python
data1.dtypes
```

```python
from sklearn.cluster import KMeans
X = data1.drop([0],axis=1)
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=2,random_state=42)
y = km.fit_predict(X_scaled)

plt.scatter(X_scaled[:,0],X_scaled[:,1],c=y)
plt.show()
```

```python
plt.scatter(X_scaled[:,0],X_scaled[:,1],c=y)
plt.show()
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, init='random', random_state=0)
X_t = tsne.fit_transform(X_scaled)
print("Количество компонент: 2")
plt.scatter(X_t[:,0],X_t[:,1])
plt.show()
    
```

```python
tsne = TSNE(n_components=3, init='random', random_state=42)
X_t3 = tsne.fit_transform(X_scaled)
from mpl_toolkits.mplot3d import Axes3D          # Тут попробовал в 3D, но получилось менее наглядно
fig = plt.figure(1, figsize=(6, 5))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)

plt.cla()
ax.scatter(X_t3[:, 0], X_t3[:, 1], X_t3[:, 2])
```
