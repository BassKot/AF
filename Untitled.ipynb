```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
%matplotlib inline
```

Погладим на данные

```python
data = pd.read_csv('task2_1_',sep='\t',header=None)
pd.set_option('display.max_columns',188)
data.head(10)
```

Много пропусков, надо их заменить. Например на '-1', т.к. "0" присутствуют в данных и так. Кроме того, колонки 4,11,14,16,48,122 похожи на таймстемп, значит их можно преобразовать в datetime, но в них тоже есть NA, это вызовет некоторые проблемы

```python
list_t = [4,11,14,16,48,122]
for i in list_t:
    data[i] = pd.to_datetime(data[i], unit='s',errors='ignore')
    
data.head()
```

Действительно, замечательно преобразовывается в datetime, пока непонятно как включать дату в классификацию. Можно конечно предположить, учитывая, что 13 колонка - это разность 16й и 14й то есть длительность сессии, что у роботных пользователей сессии должны быть короче по сравнению с реальными людьми. Но это пока будет просто предположение.

```python
delta = (data[16]-data[14])
delta1 = pd.Series(delta.values.astype(np.int64)//10**9)
print((delta1==data[13]).value_counts())
```

То есть пока мы получили таймстемповые колонки (11,48,122) с NA строками.Поэтому можно заменить NA в них на 0 и преобразовать опять в datetime, где будет дата 1970.1.1

```python
l = [11,48,122]
for col in l:
    data[col].replace('-',0, inplace=True)
    data[col] = pd.to_datetime(data[col], unit='s',errors='ignore')
data.head(10)    
```

```python
data.replace('-',-1, inplace=True)
data.head()
```

Посчитаем количество уникальных значений во всех признаках

```python
for col in data.columns:
    print(col,data[col].nunique())
```

Итак, у нас колонки с одним значением: 2,10,9,21,20,24,23,31,78-80,100-102,139,159,170,174 Которое очевидно не несет смысла, поэтому можно удалить стобцы

```python
data1 = data.loc[:,(data != 0).any(axis=0)]
data1.shape
```

Еще заметим, что столбцы 7,67,68,69,86,87,89,108,109,124,151 - имеют по 102 уникальному значению, а столбцы 126,143,144 - по 101. Это с учетом того, что мы заменили NA на -1, итого по 101 и 100 значений соответственно


Теперь посмотрим корреляцию Пирсона по признакам. Причем для df с датами и для df с таймстемпом. Чтобы определить, с кем работать удобнее, ну и сократить еще размерность признаков естественно.

```python
corr = data1.corr()
corr_triu = corr.where(~np.tril(np.ones(corr.shape)).astype(np.bool))
corr_triu = corr_triu.stack()
corr_triu.names = 'Pearson Correlation Coefficient'
corr_triu.index.names = ['Col1','Col2']
corr_triu[abs(corr_triu)>0.3].to_frame()
```

```python
data = pd.read_csv('task2_1_',sep='\t',header=None)
data.replace('-',-1, inplace=True)
corr = data.corr()
corr_triu = corr.where(~np.tril(np.ones(corr.shape)).astype(np.bool))
corr_triu = corr_triu.stack()
corr_triu.names = 'Pearson Correlation Coefficient'
corr_triu.index.names = ['Col1','Col2']
corr_triu[abs(corr_triu)>0.3].to_frame()
```

Видно, что в dataframe с таймстемпом, корреляция с другими фичами более видна. Еще можно учесть, что в дальнейшем будем использовать один из алгоритмов кластеризации (ну тут очевидно, что их надо использовать, т.к у нас задача unsupervised на 2 класса по условию), он с датами будет работать явно хуже, чем с int фичами. Поэтому я все таки выберу df без преобразования времени. Так что удаляем колонки 4,14,16 так как они не несут какой то эксклюзивной информации.

```python
data = data.drop([4,14,16], axis=1)
```

```python
data.shape
```

Кроме того удалим столбцы, где все значения одинаковые

```python
data1 = data.loc[:,(data != 0).any(axis=0)]
data1.shape
```

```python
corr = data1.corr()  # Еще разок посмотрим корреляцию
corr_triu = corr.where(~np.tril(np.ones(corr.shape)).astype(np.bool))
corr_triu = corr_triu.stack()
corr_triu.names = 'Pearson Correlation Coefficient'
corr_triu.index.names = ['Col1','Col2']
corr_triu[abs(corr_triu)>0.3].to_frame()
```

Нк и удалим 0 столбец - это хэш пользователя, он нам ни к чему. В результате получим таблицу фичей.

```python
X = data1.drop([0],axis=1)
```

Далее, так как масштаб признаков разный, нам нужно применить какой нибудь Scaler. Пусть будет StandardScaler

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Далее сначала применим KMeans, поглядим, что получится, а потом DBSCAN. Иерархическая кластеризация не подойдет, потому что количество классов в общем то известно, а она удобнее применяется, если нужно определить количество кластеров.

Плюс, сначала попробуем алгоритмы кластеризации на X_scaled. А потом попробуем снизить размерность данных с помощью TSNE, т.к в отличие от PCA он учитывает и нелинейные зависимости признаков.

```python
km = KMeans(n_clusters=2, init='k-means++', n_jobs=-1)
db = DBSCAN(n_jobs=-1)
y_km = km.fit_predict(X_scaled)
y_db = db.fit_predict(X_scaled)
```

```python
from sklearn.manifold import TSNE
tsne_2 = TSNE(n_components=2, init='random', random_state=42)
tsne_3 = TSNE(n_components=3, init = 'random', random_state=42)
X_t2 = tsne_2.fit_transform(X_scaled)
X_t3 = tsne_3.fit_transform(X_scaled)
```

```python
y_km
```

```python
plt.figure(figsize=(12,10))
plt.scatter(X_t2[:, 0], X_t2[:, 1], c=y_km, 
            edgecolor='none', alpha=0.7, s=40,
            cmap=plt.cm.get_cmap('nipy_spectral', 13))
plt.colorbar()
plt.title('T-SNE projection_2_clusters by KMeans')
```

```python
plt.figure(figsize=(12,10))
plt.scatter(X_t2[:, 0], X_t2[:, 1], c=y_db, 
            edgecolor='none', alpha=0.7, s=40,
            cmap=plt.cm.get_cmap('nipy_spectral', 13))
plt.colorbar()
plt.title('T-SNE projection_N_clusters by DBSCAN')
```

Видно, что DBSCAN выделил больше 2 классов, чего и следовало ожидать, так как он работает на плотности фич
